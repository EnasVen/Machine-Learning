# Naive Bayes - 鐵達尼號數據

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
```

```python
data = pd.read_csv("./dataset/titanic/train.csv" )
data.head()
```

```python
data.shape
```

```python
data['Sex'] = pd.Categorical(data['Sex']).codes
data['Embarked'] = pd.Categorical(data['Embarked']).codes
```

```python
data.head()
```

```python
data=data[["Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]].dropna(axis=0, how='any')
data.head()
```

```python
data = data.loc[:,["Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]]
data.head()
```

```python
data.shape
```

```python
X_train, X_test = train_test_split(data, test_size=0.8)

model = GaussianNB()

# Train classifier
model.fit(
    X_train[["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]].values,
    X_train["Survived"]
)
y_pred = model.predict(X_test[["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]])

# Print results
print("Number of mislabeled points out of a total {} points : {}, performance {:05.2f}%"
      .format(
          X_test.shape[0],
          (X_test["Survived"] != y_pred).sum(),
          100*(1-(X_test["Survived"] != y_pred).sum()/X_test.shape[0])
))

from sklearn.metrics import accuracy_score, confusion_matrix
accuracy = accuracy_score(X_test["Survived"], y_pred)
num_correct_samples = accuracy_score(X_test["Survived"], y_pred, normalize=False)

print('number of correct sample: {}'.format(num_correct_samples))
print('accuracy: {}'.format(accuracy))
```
